```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service as ChromeService
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.common.by import By
from bs4 import BeautifulSoup
import pandas as pd
```

```python
options=webdriver.ChromeOptions()
options.add_experimental_option('prefs',{
    "download.default_directory":"D:\PDF",
    "download.prompt_for_download":False,
    "plugins.always_open_pdf_externally":False
})
```

```python
df=pd.read_csv('file1.csv')
# df.head()
```

```python
part2_data=[]
driver= webdriver.Chrome(service=ChromeService(ChromeDriverManager().install()),options=options)
for i in range(len(df)):
    print(i)
    a={}
    a['Name']=df['name'][i]
    a['url']=df['url'][i]
    a['price']=df['price'][i]
    a['rating']=df['rating'][i]
    a['nReviews']=df['nReviews'][i]
    driver.get(df['url'][i])
    page_source=driver.page_source
    soup = BeautifulSoup(page_source, 'html.parser')

    Add_info_table=soup.find(id='productDetails_detailBullets_sections1')
    if Add_info_table==None: 
        continue
    trs=Add_info_table.findAll('tr')
    for tr in trs:
        if 'ASIN' in tr.text:
            req_tr=tr
            break
    a['ASIN']=req_tr.find('td').text
    
    tech_table=soup.find(id='productDetails_techSpec_section_1')
    if tech_table==None:
        continue
    trs=tech_table.findAll('tr')
    for tr in trs:
        if 'Manufacturer' in tr.text:
            req_tr=tr
            break
    a['Manufacturer']=req_tr.find('td').text[18:]

    feature_bullets=soup.find(id='feature-bullets')
    if feature_bullets==None:
        continue
    features = feature_bullets.find_all('span',class_='a-list-item')
    feature_list=[]
    for feature in features:
        feature_list.append(feature.text)
    a['description']=feature_list
    part2_data.append(a)
driver.close()
```

```python
df=pd.DataFrame(part2_data)
df.to_csv('file2.csv')
```
